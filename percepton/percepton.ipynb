```python
import numpy as np
```

## Definindo Funções e Variáveis

```python
# Taxa de aprendizado
alfa = 0.7
```

```python
# Bias
bias = 1
```

```python
# W
pesos_sinapticos = np.zeros(3)
```

```python
def degrau(valor):
    if valor >= 0: return 1
    return 0
```

```python
def soma(entrada_1, entrada_2):
    somatoria = 0
    # Primeira entrada
    somatoria += entrada_1 * pesos_sinapticos[0]
    somatoria += entrada_2 * pesos_sinapticos[1]
    somatoria += bias * pesos_sinapticos[2]

    return somatoria
```

```python
def erro(desejado, saida):
    return desejado - saida
```

```python
def ajuste_sinapses(erro, entrada_1, entrada_2):
    pesos_sinapticos[0] = pesos_sinapticos[0] + (alfa * erro * entrada_1)
    pesos_sinapticos[1] = pesos_sinapticos[1] + (alfa * erro * entrada_2)
    pesos_sinapticos[2] = pesos_sinapticos[2] + (alfa * erro * bias)
```

```python
def print_resultado(a, b, c, d, e):
    print(f"*** Época {a} ***")
    print(f"Entrada 1: {b}")
    print(f"Entrada 2: {c}")
    print(f"Saída: {d}")
    print(f"Erro: {e}\n")
```

## Treinar e testar o Percepton

```python
X = np.array([[0, 0], [0,1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
epocas = 1000
total_erros = 0
registro = 0

for i in range(1, epocas+1):
    # Chama função de somatória
    S = soma(X[registro][0], X[registro][1])
    # Chama função de ativação
    s = degrau(S)
    # Calcula o erro
    err = erro(y[registro], s)
    # Ajusta pesos sinápticos
    ajuste_sinapses(err, X[registro][0], X[registro][1])
    # Printa os resultados
    print_resultado(i, X[registro][0], X[registro][1], s, err)

    total_erros += err
    registro += 1
    registro = registro % 4
```
